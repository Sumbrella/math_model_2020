```python
import math
import pandas as pd
import numpy as np
N = 40
```

```python
X = pd.read_csv("data/X.csv", index_col=0)
X
```

```python
subV = [[0 for _ in range(N)] for _ in range(N)]
for i in range(N):
    for j in range(N):
        if X.iat[i, j] > 1:
            subV[i][j] = [0 for _ in range(np.int(X.iat[i, j]) - 1)]
subV = pd.DataFrame(subV, index=np.arange(1, N + 1), columns=np.arange(1, N + 1))
subV[subV==0] = None
subV
```

```python
subV = np.array(subV).tolist()
V = np.zeros(N)
```

```python
C = np.zeros(N)
```

```python
def openC(sections):
    for section in sections:
        C[section - 1] = 1

openC([2, 5, 8, 11, 14, 16, 21, 25, 31, 36])
```

```python


def getSafeValue(distance):
    return 1 / (1 + math.exp(distance))


def addSafeValue(x):
    global V
    global subV
    if C[x] > 0:
        V[x] += 1
        for y in range(N):
            if subV[x][y] is not None:
                d = len(subV[x][y])
                V[y] += getSafeValue(d + 1)
                for z in range(d):
                    subV[x][y][z] += getSafeValue(z + 1)  # 子节点增加安全值。

            if subV[x][y] is not None:
                d = len(subV[x][y])
                for z in range(d):
                    subV[y][x][z] += getSafeValue(d - z)
                    
```

```python
for i in range(N):
    addSafeValue(i)
```

```python
pd.DataFrame(subV).to_csv("data/method_2_subV.csv")
pd.DataFrame(V).to_csv("data/method_2_V.csv")
```
